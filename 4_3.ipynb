```python
def stepfunction(x):
  if x <= 0:
    return 0
  elif x > 0:
    return 1

def p_nand(x1, x2):
  w1, w2, wb = 0.5, 0.5, 1
  temp = x1 * w1 + x2 * w2 + wb
  return stepfunction(temp)

def xor(x1, x2):
  output = p_nand(p_nand(x1, p_nand(x1, x2)), p_nand(x2, p_nand(x1, x2)))
  return output

print(xor(0, 1))
print(xor(1, 1))
```
