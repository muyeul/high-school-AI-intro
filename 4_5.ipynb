# [STEP 1] 키와 몸무게를 배열에 저장하기

```python
from sklearn.metrics import mean_squared_error
  
Y_true = [49.35, 56.65, 59.76, 62.28, 64.42, 66.31, 68.01, 69.56, 70.99] 
x = [3.29, 5.37, 6.08, 6.64, 7.10, 7.51, 7.88, 8.21, 8.52]  
```

# [STEP 2] 몸무게를 이용하여 키를 예측하는 함수 작성하기

```python
def pred(x):
  n = len(x)
  y_pred = []

  for i in range(0, n):
    y = 4.3*x[i] + 34
    y_pred.append(y)
  return y_pred

Y_bar = pred(x)
```

# [STEP 3] MSE( ) 함수 작성하기

```python
def func_mse(Y_true, Y_bar):
  n = len(Y_true) 
  summation = 0
  for i in range (0, n):
    difference = Y_true[i] - Y_bar[i]  
    squared_difference = difference**2  
    summation = summation + squared_difference  
  MSE = summation/n  
  return MSE
```

```python
func_mse(Y_true, Y_bar)
```

# [STEP 4] scikit-learn의 mean_squared_error(Y_true, Y_bar) 함수 호출하기

```python
mean_squared_error(Y_true,Y_bar)
```
