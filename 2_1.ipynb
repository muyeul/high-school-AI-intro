# [STEP 8] 선그래프 그리기

```python
import matplotlib.pyplot as plt                                 # 그래프 출력을 위한 모듈 호출
plt.figure(figsize=(12,4))
plt.title('각 자치구별 CCTV 개수', fontsize=16)                 # 그래프 제목 
plt.plot(file_cctv['구분'],file_cctv['총계'])                   # 그래프를 그리기 위한 X,Y축 데이터 설정
plt.show()
```

# [STEP 9] 막대그래프 그리기

```python
import matplotlib.pyplot as plt                                        # 그래프 출력을 위한 모듈 호출
plt.figure(figsize=(12,4))
plt.title('각 자치구별 쓰레기무단투기용 CCTV 개수', fontsize=16)       # 그래프 제목 
plt.bar(file_cctv['구분'],file_cctv['쓰레기\n무단투기'],color='green') # 그래프를 그리기 위한 X,Y축 데이터 설정
plt.show()
```

# [STEP 10] 박스플롯 그리기

```python
import matplotlib.pyplot as plt

plt.title('각 기관별 CCTV 총계', fontsize=16)                           # 그래프 제목 
plt.boxplot(file_cctv['총계'], labels=['CCTV'])                         # 박스플롯 그래프 그리기
plt.show()
```

# [STEP 11] 데이터의 평균값 구하기

```python
import numpy as np                                                       # 수치 계산을 위한 numpy 모듈 호출
avg = np.mean(file_cctv['총계'])                                         # CCTV 총계의 평균값 구하기
print('CCTV개수의 평균값: ',avg)
```

# [STEP 12] 데이터의 중앙값, 최솟값, 최댓값 구하기

```python
import numpy as np                                                       # 수치 계산을 위한 numpy 모듈 호출
med = np.median(file_cctv['총계'])                                      
print('CCTV개수의 중앙값:',med)                                          # CCTV 총계의 중앙값 구하기
print('CCTV개수의 최솟값:',min(file_cctv['총계']))                       # CCTV 총계의 최솟값 구하기
print('CCTV개수의 최댓값:',max(file_cctv['총계']))                       # CCTV 총계의 최댓값 구하기
```
