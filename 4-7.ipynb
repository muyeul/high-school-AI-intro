# [STEP 1] 라이브러리 호출하기

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

# [STEP 2] 데이터를 읽고 x축과 y축에 사용할 열 지정하기 

```python
df = pd.read_csv('/gdrive/MyDrive/EBS/arcticGlacier.csv', sep=',', 
                 header=0)
```

```python
X = df['year']
Y = df['area']
```

# [STEP 3] 비용을 계산하는 함수 작성하기

```python
def cost_func(W, X, Y):

    hypothesis = W*X + 102.92   
    cost = tf.reduce_mean(tf.square(hypothesis-Y))

    return cost
```

# [STEP 4] cost값을 저장하는 리스트 작성하기

```python
cost_list = []
W_range = np.linspace(-3, 3, num=1000)
for feed_W in W_range:

    curr_cost = cost_func(feed_W, X, Y)
    cost_list.append(curr_cost)
```

# [STEP 5] 리스트에서 가장 작은 W값 찾아 출력하기

```python
min_index = cost_list.index(min(cost_list))
optimal_W = W_range[min_index]

print("Optimal W: %s" % optimal_W)
```

# [STEP 6] 시각화하기

```python
plt.title('Cost and Weight')
plt.plot(W_range, cost_list)
plt.axvline(optimal_W, color='red')
plt.xlabel('Weights')
plt.ylabel('Cost')
plt.show()
```
