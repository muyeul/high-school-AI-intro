# [STEP 11] 꽃 이미지 출력하기

```python
img2=cv2.imread('/gdrive/MyDrive/EBS/flower.jpg')
img2=cv2.resize(img2,(677,387), interpolation=cv2.INTER_LINEAR)

cv2_imshow(img2)
```

#[STEP 12] 펭수 이미지와 꽃 이미지를 합하기

```python
img1=cv2.imread(path)
img=img1+img2

cv2_imshow(img)
```

#[STEP 13] OpenCV의 add( )를 사용하여 이미지 결합하기

```python
img1=cv2.imread(path)
img=cv2.add(img1,img2)
cv2_imshow(img)
```

# [STEP 14] 이미지 이동시키기

```python
img=cv2.imread(path)
r,c=img.shape[0:2]
M=np.float32([[1,0,100],[0,1,100]])
new_img=cv2.warpAffine(img,M,(c,r))
cv2_imshow(new_img)
```

#[STEP 15] 이미지 회전시키기

```python
img=cv2.imread(path)
r,c=img.shape[0:2]
M=cv2.getRotationMatrix2D((c/2,r/2),90,1)
new_img=cv2.warpAffine(img,M,(c,r))
cv2_imshow(new_img)
```
