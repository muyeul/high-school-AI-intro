# [STEP 10] 'Age'의 결측 데이터를 평균으로 채우기

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
```

# [STEP 11] 'Embarked' 결측 데이터의 최빈값 찾기

```python
# Embarked별 개수
titanic.groupby('Embarked')['PassengerId'].count()
```

생존자 예측에 필요하지 않은 컬럼은 PassengerId, Name, Cabin, Ticket

# [STEP 12] 'Embarked'의 결측 데이터 채우기

```python
titanic['Embarked'] = titanic['Embarked'].fillna('S')
titanic.isna().sum()
```
