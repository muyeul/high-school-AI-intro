#[STEP 1] 입력한 문장의 단어별 인덱스를 출력하기

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text='''
AI의 미래!, 하루가 다르게 발전하는 AI의 미래 기술은?
AI의 미래는 밝은 것일까?
좋던 싫던 우리는 AI의 시대에 살고 있다.
'''

from sklearn.feature_extraction.text import CountVectorizer      # 빈도수를 벡터로 표현하기 위한 모듈 호출
corpus = [text]
vector = CountVectorizer()
print(vector.fit_transform(corpus).toarray())                    # 코퍼스로부터 각 단어의 빈도수를 기록
print(vector.vocabulary_)                                        # 각 단어의 인덱스가 어떻게 부여되었는지를 보여 줌.
```

#[STEP 2] 입력한 문장의 단어별 빈도수 출력하기
