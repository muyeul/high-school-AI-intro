#[STEP 2]한국어 문장을 단어로 나누어 출력하기

```python
from konlpy.tag import Okt
okt=Okt()
sentence = "자연어 처리는 재미있어요! 즐겁게 배워봐요!"
print(okt.morphs(sentence))                            # 형태소
print(okt.pos(sentence))                               # 품사(part of speech)
print(okt.nouns(sentence))                             # 명사
```
