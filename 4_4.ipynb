```python
import numpy as np
import pandas as pd
import matplotlib.pylab as pl
```

```python
df = pd.read_csv('/gdrive/MyDrive/EBS/arcticGlacier.csv', sep=',', header=0)
```

```python
x = df['year']
y = df['area']
```

[STEP 2] 산포도와 추세선 그리기

```python
# 추세선을 위한 계산 - 1차원의 polynomial(다항식)을 계산하기 위한 코드
z = np.polyfit(x, y, 1)                 # (X,Y,차원) 정의
p = np.poly1d(z)                        # 1차원 다항식에 대한 연산을 캡슐화

# 그래프 그리기
pl.plot(x,y,'go')                       # 산점도 그리기 산점도를 뜻할 때 'o'라고 함.
pl.plot(x,p(x),"r--")                   # 추세선 그리기

# 방정식 계산
print( "y=%.5fx+(%.5f)"%(z[0],z[1]))
```
